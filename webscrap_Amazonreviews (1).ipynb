```python
from bs4 import BeautifulSoup as bs
#https://docs.python.org/3/howto/urllib2.html
from urllib.request import urlopen

url = "https://www.amazon.in/Apple-iPhone-11-128GB-Black/product-reviews/B07XVLW7YK/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
#returns an httpresponse object
page = urlopen(url)
print(page)
#To extract the HTML from the page, first use the 
#HTTPResponse objectâ€™s .read() method, which returns a sequence of bytes. 
#Then use .decode() to decode the bytes to a string using UTF-8:
html = page.read().decode("utf-8")
soup = bs(html, "html.parser")
print(soup.get_text())

```

```python
print(soup.prettify())
```

```python
soup.find_all("img")

```

```python
table = soup.find('table')
print(table)
```

```python
Name = soup.findAll("span", {"class" : "a-profile-name"})
print(Name)
```

```python
Name = soup.findAll("span", {"class" : "a-profile-name"}) 
# find <span> element with attributes class = a-profile-name

# Extract name from span tag
Reviewer = []
for i in range(2,len(Name)):     
# Loop is initiated from 2 because we have to exclude the 
#Top Positive and Top Critical Review which otherwise will get repeated.
    Reviewer.append(Name[i].get_text())
    
print(Reviewer)
```

```python
Product_Rating = soup.findAll("i", {"class" : "review-rating"})
print(Product_Rating)
```

```python
Product_Rating = soup.findAll("i", {"class" : "review-rating"})     # find <i> element with attributes class = review-rating


Rating = []
for i in range(2,len(Product_Rating)):     # Loop is initiated from 2 because we have to exclude the Top Positive and Top Critical Review which otherwise will get repeated.
    Rating.append(Product_Rating[i].get_text())
    
print(Rating)
```

```python
Review_Title = soup.findAll("a", {"class" : "review-title-content"})     # find <a> element with attributes class = review-title-content
print(Review_Title)
```

```python
Review_Title = soup.findAll("a", {"class" : "review-title-content"})     # find <a> element with attributes class = review-title-content

# Extract name from span tag
Review_Desc = []
for i in range(0,len(Review_Title)):
    Review_Desc.append(Review_Title[i].get_text())

# We will remove the '\n' from before and after of every Review Title
Review_Desc[:] = [i.lstrip('\n').rstrip('\n') for i in Review_Desc]

print(Review_Desc)
```
