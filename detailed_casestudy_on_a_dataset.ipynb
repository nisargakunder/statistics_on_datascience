```python
from scipy.stats import norm
avg_hour_spent=  [4  ,   3   ,   5,   2   ,  5  ,  6  ,  3  ,  3  ,  4  ,  4 , 3  , 6 ,  7  , 2  , 2 ,  2  , 7  , 18 , 2 , 4  , 2   ,  2   ,  2    , 2   ,  5 ,  3  ,   6 ,    2    , 2  ,   2]
def npp1(data):
 p=[]
 t=[]
 data=np.sort(np.array(data))
 p=[(i-0.5)/len(data) for i in range(1,len(data)+1)]
 t=[norm.ppf(i,np.mean(data),np.std(data)) for i in p]
 plt.plot(data,t,"ro",data,data)
 plt.show()
npp1(avg_hour_spent)
```

```python
#clearly from the graph 18 hours is the outlier
```
